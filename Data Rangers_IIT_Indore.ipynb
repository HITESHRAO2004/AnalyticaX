```python
import pandas as pd
import numpy as np
```

```python
train= pd.read_csv('/content/training_set_features.csv')
TL= pd.read_csv('/content/training_set_labels.csv')
Test= pd.read_csv('/content/test_set_features.csv')
```

```python
R = pd.merge(train, TL, on='respondent_id', how='left')
```

```python
R
```

```python
R.info()
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()

# Iterate through each column in the DataFrame
for column in R.columns:
    if R[column].dtype == 'object':  # Check if the column is of type 'object'
        # Fit label encoder and transform values of the column
        R[column] = label_encoder.fit_transform(R[column])
        Test[column] = label_encoder.fit_transform(Test[column])
```

```python
R.info()
```

```python
import lightgbm as lgb
from sklearn.model_selection import train_test_split

columns = R.columns
columns =  [col for col in columns if col not in ['h1n1_vaccine','seasonal_vaccine']]
features = columns

X = R[columns]
# Split data into training and testing sets
X_train, X_test, y1_train, y1_test, y2_train, y2_test = train_test_split(X, R['h1n1_vaccine'], R['seasonal_vaccine'], test_size=0.2, random_state=42)
```

```python
from sklearn.metrics import roc_auc_score
```

```python
# Initialize LGBMRegressor with early stopping
lgbm_regressor = lgb.LGBMRegressor(objective='regression', num_leaves=31, learning_rate=0.05, n_estimators=100,
                                    early_stopping_rounds=10)

# Fit the model to the training data for the first target variable
lgbm_regressor.fit(X_train, y1_train, eval_set=[(X_test, y1_test)], eval_metric="roc_auc")

# Predict probabilities for the test data for the first target variable
y1_pred = lgbm_regressor.predict(X_test, num_iteration=lgbm_regressor.best_iteration_)
Test['h1n1_vaccine']=lgbm_regressor.predict(Test[features])


# Fit the model to the training data for the second target variable
lgbm_regressor.fit(X_train, y2_train, eval_set=[(X_test, y2_test)], eval_metric="roc_auc")

# Predict probabilities for the test data for the second target variable
y2_pred = lgbm_regressor.predict(X_test, num_iteration=lgbm_regressor.best_iteration_)
Test['seasonal_vaccine']=lgbm_regressor.predict(Test[features])
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc
fpr1, tpr1, _ = roc_curve(y1_test, y1_pred)
roc_auc1 = auc(fpr1, tpr1)

fpr2, tpr2, _ = roc_curve(y2_test, y2_pred)
roc_auc2 = auc(fpr2, tpr2)

# Plot ROC curve
plt.figure()
plt.plot(fpr1, tpr1, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc1)
plt.plot(fpr2, tpr2, color='blue', lw=2, label='ROC curve (area = %0.2f)' % roc_auc2)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()
```
